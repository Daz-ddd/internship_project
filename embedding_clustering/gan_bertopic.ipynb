```python
from transformers import pipeline
from bertopic import BERTopic

embedding_model = pipeline(
    "feature-extraction", model="GanymedeNil/text2vec-large-chinese")

topic_model = BERTopic(embedding_model=embedding_model,
                       calculate_probabilities=True, nr_topics=30, verbose=True)
```

```python
import pandas as pd
import numpy as np
import json

with open('data/qa_data.json', 'r') as f:
    qa_list = json.load(f)

questions_list = [qa_list[i]['question'] for i in range(len(qa_list))]
questions_list
```

```python
topics, probabilities = topic_model.fit_transform(questions_list)
```

```python
topic_model.save("model")
```

```python
# 展示所有主题
t_document = topic_model.get_document_info(questions_list)
t_document
```

```python
t_document.Topic.value_counts()
```

```python
result = pd.DataFrame()

for i in range(-1, 29):
    temp = t_document[t_document['Topic'] == i]['Document'].values
    result = pd.concat([result, pd.DataFrame(
        temp, columns=['cluster_' + str(i)])], axis=1)


result.to_csv('./clustered_data/bertopic_cluster.csv',
              index=False, encoding='utf-8')
```
